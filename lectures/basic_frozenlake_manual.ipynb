```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
```

```python
# create the FrozenLake -environment
env = gym.make("FrozenLake-v1", is_slippery=False, render_mode="rgb_array")
```

You can use the input to move ESTEBAN the ELF around the ice.

```python
# before attempting to visualize the environment, a reset() is required
env.reset()

# clear the output below
clear_output(wait=True)

# show the current frame of the game by using Matplotlib
plt.imshow(env.render())
plt.show()

# map the keys according to FrozenLake documentation
# 0: Move left, 1: move down, 2: move right, 3: move up
action_map = {"w": 3, "s": 1, "a": 0, "d": 2}

# let's make the game loop (infinite loop)
while True:
    action_key = input("Move (WASD) [use 'q' to exit]: ").strip().lower()

    # let's make an exit condition for the user
    if action_key == "q":
        break

    # if use gave a correct movement direction, move the character
    # and update the current screen
    if action_key in action_map:
        # convert the WASD into the environment action
        action = action_map[action_key]
    
        # get the observation, reward etc. from the environment => MOVE THE ELF
        observation, reward, done, truncated, info = env.step(action)

        # clear the output
        clear_output(wait=True)

        # show the current frame of the game by using Matplotlib
        plt.imshow(env.render())
        plt.show()

        # if we either fall into a hole or find the GIFT
        if done:
            print(f"Game over! Reward: {reward} - Resetting...")
            time.sleep(3)

            # game over, reset the environment
            env.reset()

            # clear the output
            clear_output(wait=True)

            # show the current frame of the game by using Matplotlib
            plt.imshow(env.render())
            plt.show()


# close the environment if q was pressed
env.close()
```
