### Let's make the CPU play the game RANDOMLY

**Note: there's no AI or RL agent here, just random directions used while playing.**

```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output

# create the FrozenLake -environment
env = gym.make("FrozenLake-v1", is_slippery=False, render_mode="rgb_array")
```

```python
# before attempting to visualize the environment, a reset() is required
env.reset()


# run the game only for 100 steps
# don't use infinite loop because we wouldn't have a exit condition
for _ in range(100):
    # get a random action between 0-3 (left, down, right, up)
    random_action = env.action_space.sample()

    # get the observation, reward etc. from the environment => MOVE THE ELF
    observation, reward, done, truncated, info = env.step(random_action)

    # clear the output
    clear_output(wait=True)

    # show the current frame of the game by using Matplotlib
    plt.imshow(env.render())
    plt.show()

    # if we either fall into a hole or find the GIFT
    if done:
        print(f"Game over! Reward: {reward} - Resetting...")
        time.sleep(3)

        # game over, reset the environment
        env.reset()

        # clear the output
        clear_output(wait=True)

        # show the current frame of the game by using Matplotlib
        plt.imshow(env.render())
        plt.show()

    # between the steps, add a slowdown
    # so we can actually see what's happening
    # time.sleep(0.35)


# close the environment if q was pressed
env.close()
```
